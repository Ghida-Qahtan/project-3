## 2- Clean and extract data:

```python
# define List of data's features
name=[]
city=[]
Condo_hotel=[]
review_badge=[]
review_titel=[]
review_number=[]
cc=[]
Favorite_partner=[]
Describion=[]
price=[]
type_hotels=[]
#name_price=[]
#space=[]
#bathroom=[]
#rooms=[]
bad_number=[]
dual_bad=[]
odd_bad=[]
bad_config=[]
left_rooms=[]
breakfast=[]
#prepaid=[]
#Free_cancellation=[]
taxes=[]
Duration=[]
persons=[]

stars=[]
#location=[]

list_of_Name_of_col=['name','city','Condo_hotel','review_badge','review_titel','review_number','cc',
'price','Duration','persons','stars','taxes','Favorite_partner','type_hotels']
```

```python
#extract each data's features than clean it than save it to its list
for i in range (0,len(item_lst)):
    for j in range (0,len(item_lst[i])):
# name
        name.append(item_lst[i][j].find('span',attrs={'class':"sr-hotel__name"}).text.replace('\n',''))
# city
        city.append(item_lst[i][j].find('a',attrs={'class':"bui-link"}).text.replace('\n','').replace(' on map',''))
# Condo_hotel
        try:  # Use try-except because some of the data is missing and will cause an error "out of range"
            Condo_hotel.append(item_lst[i][j].find(attrs={'class':"bui-badge bh-property-type bh-property-type--constructive-dark"}).text.replace('\n',''))
        except:
            Condo_hotel.append("null")

# review_badge
        try:
            review_badge.append(item_lst[i][j].find('div',attrs={'class':"bui-review-score__badge"}).text.replace(' ',''))
        except:
            review_badge.append('None')
# review_titel
        try:
            review_titel.append(item_lst[i][j].find('div',attrs={'class':"bui-review-score__title"}).text.replace(' ',''))
        except:
            review_badge.append('None')        
# review_number
        try:
            review_number.append(item_lst[i][j].find('div',attrs={'class':"bui-review-score__text"}).text.replace(' ',''))
        except:
            review_badge.append('None')
# cc
        try:  # Use try-except because some of the data is missing and will cause an error "out of range"
            cc.append(item_lst[i][j].find(attrs={'class':"no_cc_02"}).text.replace('\n','').text)
        except:
            cc.append("null")
        
# left_rooms
        try: 
            left_rooms.append(item_lst[i][j].find(attrs={'class':"only_x_left sr_rooms_left_wrap"}).text.replace('\n','').replace('Only ','').split(' like')[0])
        except:
            left_rooms.append("available")
# name
#        name_price.append(item_lst[i][j].find(attrs={'class':"sr-hotel__name"}).text.replace('\n',''))

# price
        try: 
            price.append(item_lst[i][j].find(attrs={'class':"bui-price-display__value prco-inline-block-maker-helper"}).text.replace('\n','').replace('\xa0',' '))
        except:
            price.append("null")
# bad_number
        try:  # Use try-except because some of the data is missing and will cause an error "out of range"
            bad_number.append(item_lst[i][j].find('div',attrs={'class':'c-beds-configuration'}).text.replace('\n','').split(' beds')[0])
        except:
            bad_number.append("null")
# bad_config
        try:  # Use try-except because some of the data is missing and will cause an error "out of range"
            bad_config.append(item_lst[i][j].find('div',attrs={'class':'c-beds-configuration'}).text.replace('\n','').split(' beds')[1].replace('(','').replace(')',''))
        except:
            bad_config.append("null")
                
# Duration
        try:  # Use try-except because some of the data is missing and will cause an error "out of range"
            Duration.append(item_lst[i][j].find(attrs={'class':"bui-price-display__label prco-inline-block-maker-helper"}).text.split(",")[0])
        except:
            Duration.append("null")

# persons
        try:  # Use try-except because some of the data is missing and will cause an error "out of range"
            persons.append(item_lst[i][j].find(attrs={'class':"bui-price-display__label prco-inline-block-maker-helper"}).text.split(",")[1])
        except:
            persons.append("null")
#Favorite_partner        
        try:  # Use try-except because some of the data is missing and will cause an error "out of range"
            Favorite_partner.append(item_lst[i][j].find(attrs={'class':"sr-hotel__title-badges"}).text.replace('\n',''))
        except:
            Favorite_partner.append("null")
# taxes
        taxes.append(item_lst[i][j].find(attrs={'class':"prd-taxes-and-fees-under-price prco-inline-block-maker-helper blockuid-"}).text.replace('\n','').replace('\xa0',''))
    
# type_hotels
        try:  # Use try-except because some of the data is missing and will cause an error "out of range"
            type_hotels.append(item_lst[i][j].find(attrs={'class':"bui-badge bh-property-type bh-property-type--constructive-dark"}).text.replace('\n',''))
        except:
            type_hotels.append("null")        
```
